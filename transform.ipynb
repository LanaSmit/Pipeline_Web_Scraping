```python
import polars as pl
# Could have use polars
df = pl.read_csv("netflix_top10.csv")
df.head()
```

```python
#normalize the column names
df.columns = [col.lower().replace(" ","_") for col in df.columns]
```

```python
#remove the numbers from the "title" column
df = df.with_columns(
    pl.col("title").str.replace(r"^\d+", "").alias("title")
)
```
